## 1. Linear Regression
공부 시간에 따른 시험 점수


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset.head()
```

```python
X = dataset.iloc[:,:-1].values 
# 처음부터 마지막 컬럼 직전까지의 데이터 [독립 변수 -> 원인]
# 2차원 배열로 나타낼 수 있다.
print(X)
# x1 = dataset['hour'].values
# print(x1)
# x1은 1차원 배열으로 머신러닝 모델의 입력으로 사용하지 못한다.
y = dataset.iloc[:,-1].values
# 마지막 컬럼 데이터 [종속 변수 -> 결과]
print(y)
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() # 객체 생성
reg.fit(X,y) # 학습 (모델 생성)
```

```python
y_pred =reg.predict(X) # X에 대한 예측 값
y_pred
```

```python
plt.scatter(X,y,color='blue') # 산점도
plt.plot(X,y_pred,color = 'green') # 선 그래프
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
print('9시간 공부했을 때 점수 :',reg.predict([[9]]))
```

```python
print('9시간 공부했을 떄 예상 점수 : ',reg.predict([[9],[8],[7]]))
```

```python
reg.coef_ # 기울기 [m]
```

```python
reg.intercept_ # y 절편 [b]
```

데이터 세트 분리

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset.head()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0) 
# 훈련 80 : 테스트 20 으로 분리
```

```python
X,len(X) # 전체 데이터 X, 개수
```

```python
X_train,len(X_train) # 훈련 세트 X 개수
```

```python
X_test, len(X_test) # 테스트 세트 X , 개수
```

```python
y , len(y)
```

```python
y_train,len(y_train)
```

```python
y_test,len(y_test)
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train) # 훈련 세트로 학습
```

### 데이터 시각화 (훈련 세트)

```python
plt.scatter(X_train,y_train,color ='b')
plt.plot(X_train,reg.predict(X_train),color ='g')
# 모델의 직선을 보여준다.
plt.title('Score by hours(train data)')
plt.xlabel('hours') # X 축 이름
plt.ylabel('score') # Y 축 이름
plt.show()
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test,y_test,color="b")
plt.plot(X_train,reg.predict(X_train),color='g')
plt.title('Score by hours (test_data)')
plt.xlabel('hours')
plt.ylabel('Score')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```


### 모델 평가

```python
reg.score(X_test,y_test) # 테스트 세트를 통한 모델 평가 (0~1)
```

```python
reg.score(X_train,y_train) # 훈련 세트를 통한 모델 평가
```

경사 하강법 (Gradient Descent)

```python
from sklearn.linear_model import SGDRegressor
# 확률적 경사 하강법
```

max_iter : 훈련 세트 반복 횟수 (Epoch 횟수)

eta0 : 학습률 (learning rate)

```python

# 지수 표기법
# le-3 : 0.001 (10^-3)
# le-4 : 0.0001 (10^-4)
# le+3 : 1000
# le+4 : 10000
# sr = SGDRegressor(max_iter=200,eta0=0.0001,random_state=0,verbose=1)

sr = SGDRegressor(max_iter=350,eta0=0.0001,random_state=0,verbose=1)
sr.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train,color ='g')
plt.plot(X_train,sr.predict(X_train),color='r')
plt.show
```

```python
sr.coef_ ,sr.intercept_
```

```python
sr.score(X_train,y_train) # 훈련 세트를 통한 모델 평가 
# 보통 훈련 세트 모델 평가 점수가 높다
```

```python
sr.score(X_test,y_test) # 테스트 세트를 통한 모델 평가
```
