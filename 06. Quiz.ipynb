```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1) 파일로부터 데이터를 불러와 결혼식 참석 인원(total), 식수 인원(reception)을 각각의 변수로 저장하시오.

```python
dataset = pd.read_csv('QuizData.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
X[:5],y[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)

```

```python
len(X)
```

훈련 세트

```python
plt.scatter(X_train,y_train,color='blue')
plt.plot(X_train,reg.predict(X_train),color='green')
plt.xlabel('total')
plt.ylabel('reception')
```

테스트 세트

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_train,reg.predict(X_train),color='g')
plt.plot(X_test,reg.predict(X_test),color='r')
```

```python
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
total = 300
y_pred = reg.predict([[total]])
print(f"결혼식 참석 인원 300명에 대한 예상 식수 인원은 {np.around(y_pred[0].astype(int))}명 입니다.")
```
